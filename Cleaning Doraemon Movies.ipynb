# Import Dependencies

```python
import pandas as pd
```

# Load Data

```python
data = pd.read_csv("doraemon_movies.csv")
```

```python
data = data.loc[:,"Movie":]
```

```python
data.head()
```

# Mengecek Nama Negara dan Tanggal Rilis

```python
isCountry = False
countryDate = []
ultiCountryDate = []
country = []
# statebreak=False
for iMovie in range(len(data)):
    countryDate = []
    date = data.loc[iMovie,"Date"].split("\n")
    # Mendapatkan semua element (Date dan Country)
    for element in date:
        # apabila element memiliki digit maka Country
        if (any(char.isdigit() for char in element))==False and isCountry==False:
            isCountry=True
            countryDate.append(element)
            country.append(element)
        # apabila element tidak memiliki digit maka bukan Country
        elif (any(char.isdigit() for char in element))==True and isCountry==True:
            isCountry=False
            countryDate.append(element)
        else:
            continue
    #     if element=="":
    #         statebreak=True
    # if statebreak==True:
    #     break
    ultiCountryDate.append(countryDate)
country=set(country) #mendapatkan semua country yang tersedia
```

```python
country=pd.DataFrame(columns=list(country)[1:],index=[0]) # membersihkan data country
```

```python
# merapikan menjadi dataframe
for idata in range(len(data)):
    for ielement in range(len(ultiCountryDate[idata])):
        if any(char.isdigit() for char in ultiCountryDate[idata][ielement])==False:
            Country = ultiCountryDate[idata][ielement]
            Date = ultiCountryDate[idata][ielement+1]
            country.loc[idata,Country]=Date
        
```

```python
country.head()
```

```python
ulti = pd.concat([data,country],axis=1)
```

```python
ulti.drop(["Date",""],axis=1,inplace=True)
```
