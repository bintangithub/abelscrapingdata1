```python
import pandas as pd
import re
```

```python
movies = pd.read_csv("marvel_movies.csv",index_col=False)
```

```python
movies = movies.loc[:,"Marvel Movies":]
```

```python
movies
```

```python
for daterelease in movies["Marvel Movies"]:
    date = re.findall(r"(\d{4})\)?",daterelease)
    print(date[0])
```

```python
for movie in movies["Marvel Movies"]:
    movie = re.findall(r"\d+\.\s([A-z\s.:’-]+[\d+]?)\s",movie)
    print(movie[0])
```

```python
re.findall(r"[A-z\s:]+",movies.loc[11,"Marvel Movies"])
```

```python
# memasukan data menjadi dataframe
dictMarvel = {
    "Movies" : [],
    "Release Year": []
             }
for movie in movies["Marvel Movies"]:
    try:
        searchedmovie = re.findall(r"\d+\.\s([A-z\s.:’-]+[\d+]?)\s",movie)
        date = re.findall(r"(\d{4})\)?",movie)
        dictMarvel["Movies"].append(searchedmovie[0])
        dictMarvel["Release Year"].append(date[0])
    except:
        dictMarvel["Release Year"].append(None)
```
